#  Grammar Of Graphics
A short Note on Grammar Of Graphics Before we proceed to How to plot interactive visualizations in Python

## Q. What exactly is grammar of Graphics?
A grammar of graphics is a framework which follows a layered approach to describe and construct visualizations or graphics in a structured manner.

## Q. Why is it useful?
A visualization involving multi-dimensional data often has multiple components or aspects, and leveraging this layered grammar of graphics helps us describe and understand each component involved in visualization — in terms of data, aesthetics, scale, objects and so on. 


## Q. Can you provide some context of Grammar of Graphics? Is it possible to do that completely in one session?
- The original grammar of graphics framework was proposed by Leland Wilkinson, which covers all major aspects pertaining to effective data visualization in detail. 
- However, here we will be using a variant of this — known as the layered grammar of graphics framework, which was proposed by Hadley Wickham, reputed Data Scientist and the creator of the famous R visualization package ggplot2.
-  The text which follows has taken its inspiration from his famous research paper [A Layered Grammar Of Graphics](http://vita.had.co.nz/papers/layered-grammar.pdf)


## Q. What do you mean by layers in a graphics
You can see there are various components in the grammar which can be used to build and describe a visualization.Hadley identified the following components or layers
 - data 
 - aesthetic Mapping 
 - geometric objects
 - scales  
 - facet specification
 - Statistical Transformation 
 - Coordinate system

 image.png

Courtsey: https://towardsdatascience.com/a-comprehensive-guide-to-the-grammar-of-graphics-for-effective-visualization-of-multi-dimensional-1f92b4ed4149#:~:text=each%20specific%20component%20in%20the%20grammar.

1. **Data**: Always start with the data, identify the dimensions you want to visualize.
2. **Aesthetics**: Confirm the axes based on the data dimensions, positions of various data points in the plot. Also check if any form of encoding is needed including size, shape, color and so on which are useful for plotting multiple data dimensions.
3. **Scale**: Do we need to scale the potential values, use a specific scale to represent multiple values or a range?
4. **Geometric objects**: These are popularly known as ‘geoms’. This would cover the way we would depict the data points on the visualization. Should it be points, bars, lines and so on?
5. **Statistics**: Do we need to show some statistical measures in the visualization like measures of central tendency, spread, confidence intervals?
6. **Facets**: Do we need to create subplots based on specific data dimensions?
7. **Coordinate system**: What kind of a coordinate system should the visualization be based on — should it be cartesian or polar?

We will be using [plotnine](https://plotnine.readthedocs.io/en/stable/) library in python,which is an implementation of a grammar of graphics in Python, it is based on ggplot2. The grammar allows users to compose plots by explicitly mapping data to the visual objects that make up the plot.

image.png

```python
  (ggplot(mtcars,                                            # data step, here we are using mtcars dataset 
          aes('wt','mpg',color='factor(gear)'))              # Aesthetics in that we are using wt and mpg columns, also we are coloring based on gear
  + geom_point()                                             # Geometric objects: point
  + xlim(0,7)+ylim(10,40)                                    # Scale Value: x limit and y limit
  + stat_summary()                                           # Statistics
  + facet_wrap('~am')                                        # Facet
  + theme_bw()
                                                            # coordinate system is cartesian
  )
```

## Some Examples for you to try

In the below three code cells, we have mentioned codes based on Grammar Of Graphics, you would have to identify which layer is being used at which step.

```python
import pandas as pd

n = 10
df = pd.DataFrame({'x': np.arange(n),
                   'y': np.arange(n),
                   'yfit': np.arange(n) + np.tile([-.2, .2], n//2),
                   'type': ['a', 'b']*(n//2)})

(ggplot(df)
 + geom_col(aes('x', 'y', fill='type'))
 + geom_point(aes('x', y='yfit', color='type'))
 + geom_path(aes('x', y='yfit', color='type'))
 + ggtitle('Visualization 1')
)
```

```python
from plotnine.data import diamonds
(ggplot(diamonds, aes('carat', 'price')) +
geom_point() +
stat_smooth(method = 'lm',color='red') +
scale_x_log10() +
scale_y_log10()+
ggtitle('Visualization 2'))
```

```python
def combine(counts, percentages):
    fmt = '{} ({:.1f}%)'.format
    return [fmt(c, p) for c, p in zip(counts, percentages)]
(ggplot(mtcars, aes('factor(cyl)', fill='factor(cyl)'))
 + geom_bar()
 + geom_text(
     aes(label=after_stat('combine(count, prop*100)'), group=1),
     stat='count',
     nudge_y=0.125,
     va='bottom',
     size=9
 )
 + facet_wrap('am')
 + ggtitle('visualization 3')
)
```

## A side note on different modules of Plotly

**plotly.express** is a high-level interface for creating easy-to-use and expressive plots. It provides a simple API for creating common chart types such as scatter, line, bar, and histogram, as well as more specialized visualizations like treemaps, sunbursts, and 3D plots. plotly.express is built on top of plotly.graph_objects and provides a streamlined interface that makes it easy to create complex visualizations with minimal code.

**plotly.graph-objects** is a lower-level interface that provides complete control over the appearance and behavior of a plot. It allows users to specify every aspect of a plot, from the color of individual points to the position of labels and annotations. plotly.graph-objects is particularly useful for creating custom visualizations that require fine-grained control over the layout and appearance of a plot.

## Basic Charts
  ### Scatter Plots and Bubble Chart
  - Scatter plots are the graphs that present the relationship between two variables in a data-set. 
  - Scatter plots with variable-sized circular markers are often known as **bubble charts**

```python
# Scatter Plots
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", symbol="species")
fig.show()
```

- In the above example scatter plot higlights relationship between sepal_lenght and sepal_width of different species, different color and symbols are used to distinguish between different species.
- This chart helps highlight relationship between sepal length and sepal width for different species

```python
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'])
fig.show()
```

Above is a bubble chart where size of the circle is in proportion of petal length, this chart gives us another dimension of visualization and ones you hover around the data you also get its petal width
Thus, we can see bubble charts helps in visualizing more dimensions compared to scatter plot

### Line Chart
A line chart or line graph, also known as curve chart, is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments

```python
df = px.data.gapminder().query("continent=='Oceania'")
fig = px.line(df, x="year", y="lifeExp", color='country')
fig.show()
```

The above Example is a gives us trends of life expectancy in continetn of Oceania from 1952 to 2007, the lines (depicting different countries) also enables year on year comparisions
