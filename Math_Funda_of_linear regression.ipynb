```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 6, 8, 10, 12])
```

```python
plt.plot(x, y)
plt.show()
```

```python
w0, b = 0.1, 1.0
lr = 0.001
epochs = 1000
error= []
```

```python
for epoch in range(epochs):
    epoch_cost, cost_w0, cost_b =  0,  0, 0
    for i in range(len(x)):
        y_pred = w0*x[i] + b
        epoch_cost += (y[i]-y_pred)**2
        
        for j in range(len(x)):
            partial_wrt_b = -2*(y[j]-(w0*x[j] + b))
            partial_wrt_w0 = (-2*x[j])*(y[j]-(w0*x[j] + b))
            
            cost_w0 += partial_wrt_w0
            cost_b += partial_wrt_b
            
        w0 = w0-lr*cost_w0
        b = b-lr*cost_b
        
        print("w0 = ", w0)
        print("b = ", b)
    error.append(epoch_cost)
```
