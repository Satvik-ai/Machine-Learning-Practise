```python
from sklearn.model_selection import cross_validate
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
X = [[1, 2], [3, 4], [1, 2], [3, 4]]
def k_fold_cv(X):
  for _ in range(2):
    cv = cross_validate()
```

```python
data = fetch_california_housing(as_frame=True)
X = data.data # Feature Matrix
y = data.target # Label Vector
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1,test_size=0.3)
```

```python
ss = StandardScaler(with_mean=True,with_std=True)
X_train_norm = ss.fit_transform(X_train)
X_test_norm = ss.transform(X_test)
```

```python
sgd =SGDRegressor(random_state=1)
param_grid = [
    {'loss':['squared_error','huber'],'penalty':['l1','l2'],'alpha':[0.1,0.01,0.001],'max_iter':[1000,2000,5000]}
]
grid_search = GridSearchCV(sgd,param_grid,cv=4)
grid_search.fit(X_train_norm,y_train)
```

```python
grid_search.score(X_test_norm,y_test)
```

```python
grid_search.best_params_
```

```python
param_grid = [
    {'alpha':[0.5,0.1,0.05,0.01,0.005,0.001],'fit_intercept':[True,False]}
]
GSCV = GridSearchCV(Ridge(),param_grid,cv=4)
GSCV.fit(X_train_norm,y_train)
```

```python
GSCV.score(X_test_norm,y_test)
```

```python
GSCV.best_params_
```

```python
def k_closest(X,p,k=1):
  min = np.inf
  D = []
  for x in X:
    dist = np.linalg.norm(x-p)
    D.append(dist)
    if dist < min:
      min = dist
      ans = x
  return ans
```

```python
X = np.asarray([[72, 69 ,82], [ 9 ,79, 99], [20 ,47, 88], [80 ,64, 49]])
p= np.asarray([[0,0,0]])
print(k_closest(X,p,1))
```

```python
def distance(x0,X,p=2):
  ans = []
  for x in X:
    Dict = 0
    for i in range(len(x)):
      Dict += (abs(x[i]-x0[i]))**p
    ans.append(Dict**(1/p))
  return ans
```

```python
X0 = np.array([1,0,0,0])
X= np.asarray([[1, 0, 0,0], [0, 1, 1,1],[1,2,0,0]])
p= 2
print(distance(X0,X,p))
```
